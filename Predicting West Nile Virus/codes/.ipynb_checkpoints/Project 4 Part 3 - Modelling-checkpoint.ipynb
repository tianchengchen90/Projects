````python
train_weather_df = pd.read_csv('../datasets/train_weather_model.csv')
````

# Model Merged Dataset Preprocessing

## Set X and y Variables/Train Test Split

````python
# Set X and y variables

X=train_weather_df.drop('WnvPresent',axis=1)
y=train_weather_df['WnvPresent']

#Train test split
X_train, X_test, y_train,y_test =train_test_split(
    X,
    y,
    stratify=y,
    test_size = 0.3,
    random_state=50,
)
````

````python
print(y_train.shape)
print(y_test.shape)
print(X_train.shape)
print(X_test.shape)
print(y_train.value_counts(normalize=True))
print(y_test.value_counts(normalize=True))
````

The shape of both post-split train and test sets were verified to ensure the they are aligned. Based on value count result, we found that proportion of negative classes is around 95% and the positive classes is around 5%, which is characteristic of imbalanced classes in this dataset.

## Standardization

````python
ss=StandardScaler()
X_train_sc=ss.fit_transform(X_train)
X_test_sc=ss.transform(X_test)
````

# Model Selection

## Base Model

We first perform a Logistic Regression with default parameters to obtain a Base Model to compare our metrics to. Sensitivity, Accuracy, Specificity, Precision, F1 Score, Train Score, Test Score and ROC AUC will be measured for each model, but for model-to-model comparisons we will be evaluating solely on ROC AUC score.

````python
# Instantiate model.
logreg = LogisticRegression(max_iter = 5_000)

# Fit model.
logreg.fit(X_train_sc, y_train)

# Generate predicted values.
preds = logreg.predict(X_test_sc)

#Generate predicted probabilities.
preds_proba = np.round(logreg.predict_proba(X_test_sc), 3)
````

````python
print('Logistic Regression')
print('Train score: {}, Test score {}'.format(round(logreg.score(X_train_sc, y_train), 4), round(logreg.score(X_test_sc, y_test), 4)))
````

````python
#metrics calculations
tn, fp, fn, tp = confusion_matrix(y_test, preds).ravel()
acc = (tp + tn) / (tn + fp + fn + tp)
misclass = 1 - acc
sens = tp / (tp + fn)
spec = tn / (tn + fp)
prec = tp / (tp + fp)
f_one_score = f1_score(y_test, preds)
print('Sensitivity:', sens)
print('Accuracy:',acc)
print('Specificity:',spec)
print('Precision:',prec)
print('f1 score:',f_one_score)
````

````python
fig, axes = plt.subplots(ncols=2, figsize=(17,5))
cm_chart = plot_confusion_matrix(logreg, X_test_sc, y_test, cmap='Blues', values_format='d', ax=axes[0]);
axes[0].title.set_text('Confusion Matrix')
roc_curve = plot_roc_curve(logreg, X_test_sc, y_test, ax=axes[1])
axes[1].plot([0, 1], [0, 1], label='baseline', linestyle='--' )
axes[1].title.set_text('ROC AUC Curve Chart')

fig.tight_layout()
````

Based on the results, we have Train score=0.947 and Test score=0.947, which is normally good result. However, the reported accuracy score for both instances are actually the baseline class distribution of the Train and Test sets.

By taking a closer look, we see that Sensitivity is 0 and Precision is NAN. From the confusion matrix above, we confirmed that this model is not even able to predict a single instance of WnvPresent (positive class). The reason is due to the high imbalanced distribution of positive vs. negative classes mentioned and currently, the data in the positive class is not sufficient in providing a good predictive model.

Thus, in our next of exploration of models, we will be using Synthetic Minority Oversampling Technique (SMOTE) which aims to balance class distribution by randomly increasing minority class by replicating the minorty observations.

##  Logistic Regression (with SMOTE)

````python
# Creating a pipeline model for SMOTE and Logistic Regression.
pipe_smote_logreg = Pipeline([
                ('sampling', SMOTE(random_state=100)),
                 ('logreg', LogisticRegression(solver='liblinear', random_state=100))
                ])

pipe_smote_logreg_params = {
    'logreg__penalty': ['l1'],
    'logreg__C': [1],
    'sampling__sampling_strategy': ['auto','minority'],
    'sampling__k_neighbors': [13],
}

gs_smote_logreg = GridSearchCV(pipe_smote_logreg, pipe_smote_logreg_params, verbose=2, n_jobs=-1)

gs_smote_logreg.fit(X_train_sc, y_train) 
````

````python
print('Logistic Regression')
print("Best params: {}".format(gs_smote_logreg.best_params_))
print('Train score: {}, Test score {}'.format(round(gs_smote_logreg.best_score_, 4), round(gs_smote_logreg.best_estimator_.score(X_test_sc, y_test), 4)))
````

````python
#metrics calculations
preds = gs_smote_logreg.predict(X_test_sc)
tn, fp, fn, tp = confusion_matrix(y_test, preds).ravel()
acc = (tp + tn) / (tn + fp + fn + tp)
misclass = 1 - acc
sens = tp / (tp + fn)
spec = tn / (tn + fp)
prec = tp / (tp + fp)
f_one_score = f1_score(y_test, preds)
roc_auc = roc_auc_score(y_test, gs_smote_logreg.predict_proba(X_test_sc)[:,1])
print('Sensitivity:', sens)
print('Accuracy:',acc)
print('Specificity:',spec)
print('Precision:',prec)
print('f1 score:',f_one_score)
````

````python
fig, axes = plt.subplots(ncols=2, figsize=(17,5))
    
cm_chart = plot_confusion_matrix(gs_smote_logreg, X_test_sc, y_test, cmap='Blues', values_format='d', ax=axes[0]);
axes[0].title.set_text('Confusion Matrix')
roc_curve = plot_roc_curve(gs_smote_logreg, X_test_sc, y_test, ax=axes[1])
axes[1].plot([0, 1], [0, 1], label='baseline', linestyle='--' )
axes[1].title.set_text('ROC AUC Curve Chart')
fig.tight_layout()
````

For this model, we are re-running Logistic Regression with SMOTE, with input variables similar to Base Model. We see improvement in this model in detecting the minority class although the Accuracy has dropped to 0.758. ROC AUC score has dropped to 0.78. Despite that, this model has accurately predicted 108 observations with WNV present, a marked increase from the Base Model.

# Hyperparameter Tuning

For XGBoost, there is an inbuilt hyperparameter `'scale_pos_weight'` that can help to take care of imbalanced class datasets like the one we have here. Thus, this will be used instead of SMOTE.

Based on the documentation, the suggested number to use is 19 since this is a 5 to 95 (1 to **19**) ratio of minority to majority class and this will be used as a base to perform hyperparameter of others like `'gamma'`, `'learning_rate'`, `'max_depth'`, `'min_child_weight'`, `'n_estimators'`, and `'subsample'`. ROC AUC was used for XGBoost and GridSearchCV evaluation.

Specific to our problem statement, our objective is to predict when and where different species of mosquitos will test positive for West Nile virus so that there is effective allocation of resources towards peventing transmission. Other than accurately predicting the number of positive classes, we also need to reduce the false negatives to a minimum number so that no regions are missed out despite sacrificing our False Positives count (focusing on Sensitivity). This will be best in ensuring the risk of WNV virus to the population if Chicago at its utmost minimum.



````python
pipe_xgb_params = {
        'xgb__gamma': [1],
        'xgb__learning_rate': [0.1],
        'xgb__max_depth': [5],
        'xgb__min_child_weight': [25],
        'xgb__n_estimators': [60],
        'xgb__scale_pos_weight': [19],
        'xgb__subsample': [1]}
    
pipe_xgb = Pipeline([    
                    ('xgb',XGBClassifier(
                        objective='binary:logistic',
                        verbosity=1, 
                        n_jobs=-1, 
                        random_state=42, 
                        eval_metric='auc', 
                        use_label_encoder =False)),
                    ])

gs_xgb = GridSearchCV(estimator=pipe_xgb, param_grid=pipe_xgb_params, cv=5, n_jobs=-1)

gs_xgb.fit(X_train_sc, y_train) 
````

````python
print('XGB')
print("Best params: {}".format(gs_xgb.best_params_))
print('Train score: {}, Test score {}'.format(round(gs_xgb.best_score_, 4), round(gs_xgb.best_estimator_.score(X_test_sc, y_test), 4)))
print()
preds = gs_xgb.predict(X_test_sc)
tn, fp, fn, tp = confusion_matrix(y_test, preds).ravel()
acc = (tp + tn) / (tn + fp + fn + tp)
misclass = 1 - acc
sens = tp / (tp + fn)
spec = tn / (tn + fp)
prec = tp / (tp + fp)
f_one_score = f1_score(y_test, preds)
print('Sensitivity:', sens)
print('Accuracy:',acc)
print('Specificity:',spec)
print('Precision:',prec)
print('f1 score:',f_one_score)
fig, axes = plt.subplots(ncols=2, figsize=(17,5))
cm_chart = plot_confusion_matrix(gs_xgb, X_test_sc, y_test, cmap='Blues', values_format='d', ax=axes[0]);
axes[0].title.set_text('Confusion Matrix')
roc_curve = plot_roc_curve(gs_xgb, X_test_sc, y_test, ax=axes[1])
axes[1].plot([0, 1], [0, 1], label='baseline', linestyle='--' )
axes[1].title.set_text('ROC AUC Curve Chart')
fig.tight_layout()
````

From the extensive hyperparemeter tuning done, we have derived the above parameters as the one that yields the model with the highest ROC AUC, given `'scale_pos_weight' = 19`.

However, we do still see 42 cases whereby the model was not able to effectively detect for WNV presence. Thus, our team decided to bias more for the positive class through tuning `'scale_pos_weight'`.


````python
pipe_xgb_params = {
        'xgb__gamma': [1],
        'xgb__learning_rate': [0.1],
        'xgb__max_depth': [5],
        'xgb__min_child_weight': [25],
        'xgb__n_estimators': [60],
        'xgb__scale_pos_weight': [63],
        'xgb__subsample': [1]}
    
pipe_xgb = Pipeline([    
                    ('xgb',XGBClassifier(
                        objective='binary:logistic',
                        verbosity=1, 
                        n_jobs=-1, 
                        random_state=42, 
                        eval_metric='auc', 
                        use_label_encoder =False)),
                    ])

gs_xgb = GridSearchCV(estimator=pipe_xgb, param_grid=pipe_xgb_params, cv=5, n_jobs=-1)

gs_xgb.fit(X_train_sc, y_train) 
````

````python
print('XGB')
print("Best params: {}".format(gs_xgb.best_params_))
print('Train score: {}, Test score {}'.format(round(gs_xgb.best_score_, 4), round(gs_xgb.best_estimator_.score(X_test_sc, y_test), 4)))
print()
preds = gs_xgb.predict(X_test_sc)
tn, fp, fn, tp = confusion_matrix(y_test, preds).ravel()
acc = (tp + tn) / (tn + fp + fn + tp)
misclass = 1 - acc
sens = tp / (tp + fn)
spec = tn / (tn + fp)
prec = tp / (tp + fp)
f_one_score = f1_score(y_test, preds)
print('Sensitivity:', sens)
print('Accuracy:',acc)
print('Specificity:',spec)
print('Precision:',prec)
print('f1 score:',f_one_score)
fig, axes = plt.subplots(ncols=2, figsize=(17,5))
cm_chart = plot_confusion_matrix(gs_xgb, X_test_sc, y_test, cmap='Blues', values_format='d', ax=axes[0]);
axes[0].title.set_text('Confusion Matrix')
roc_curve = plot_roc_curve(gs_xgb, X_test_sc, y_test, ax=axes[1])
axes[1].plot([0, 1], [0, 1], label='baseline', linestyle='--' )
axes[1].title.set_text('ROC AUC Curve Chart')
fig.tight_layout()
````

Using the same hyperparameters as the previous model, our team found that `'scale_pos_weight' = 63` was most effective in 
reducing the number of observations in the False Negative class. This can also be seen by the increase in Sensitivity. However, we do see that the number of False Positives has increased a lot. The decision and reasoning behind this tradeoff will be discussed further in Cost Benefit Analysis.

Nonetheless, this model will be chosen as our Production Model.

## Feature Importance Analysis

````python
# create table to show the predictors with the highest importance (i.e. strongest predictors of the presence of WNV)
feature_names = X.columns
coefficients = pd.DataFrame(np.squeeze(gs_xgb.best_estimator_.named_steps["xgb"].feature_importances_),
                            columns=['Importances']) 
                            
features = pd.DataFrame(np.squeeze(feature_names), columns=['Features']) 
df = pd.concat([features, coefficients], axis='columns').sort_values('Importances', ascending=False)

df.head(15)
````

**Weather Specific Features**
- The feature `'Sunrise'` ended up being a feature of high importance. This is probably due to it being a good indicator of seasonality because during summer, the days will be longer and that will correlate to an earlier sunrise. Moreover, this is probably due to a lack in some sort of time indicator in the features selected for the model. 
- For `'StnPressure'`, a rise in atmospheric pressure would generally be associated with clear weather and moderately strong winds, while a drop in atmospheric pressure would be associated with inclement weather. [source](https://www.redalyc.org/journal/5763/576366658009/html/)
- The features `'AvgSpeed'` and `'ResultSpeed'`, in relation to wind speed, contributes to virus spread by their impact on increasing the flight distance/spread of mosquitoes.
- The feature `'PrecipTotal'` is an important feature as it will inversely affect the WNV occurence. Rainfall can affect flight speed and patterns. On the other hand, low rainfall can also promote breeding as standing water becomes richer in nutrients required for their larvae survival and proliferation.
- Temperature related feature such as `'DewPoint'`, `'Tavg'`, `'Tmin'` contributes to the WNV occurence. As explored in EDA, generally, when `'DewPoint'`, `'Tavg'`, `'Tmin'` increases, the number of mosquitoes increases and in turn, WNV observations increases.

**Train Dataset Specific Features**
- Feature Engineered features such as `'cluster'` and `'Row_count'` appears in the table above. Hence, this means that our feature engineering was effective in identifying interactivity between features and engineering for stronger correlated features.
- As explored in EDA, `'Species_CULEX PIPIENS'` and `'Species_CULEX RESTUANS'` are important features for our model due to its relatively higher correlation to WNV Present, with `'Species_CULEX PIPIENS'`, with a positive correlation, having a higher importance since our model biases for the positive class.
